## Data Cleaning

Required actions:
- Investigate and remove rows with duplicated movie titles
- Convert columns to correct data types

```python
data.sample()
```

```python
data[data.duplicated(['Movie_Title','Release_Date'], keep=False)]
```

```python
data.drop_duplicates(['Movie_Title','Release_Date'], keep=False, inplace=True)
```

```python
data[data.duplicated(['Movie_Title','Release_Date'], keep=False)]
```

```python
data.duplicated(['Movie_Title','Release_Date'], keep=False).value_counts()
```

- There are 5389 rows remaining after removing the 2 duplicates

```python
data.sample()
```

```python
data['USD_Production_Budget'] = data['USD_Production_Budget'].astype(str).str.replace('\\$', "", regex=True)
data['USD_Worldwide_Gross'] = data['USD_Worldwide_Gross'].astype(str).str.replace('\\$', "", regex=True)
data['USD_Domestic_Gross'] = data['USD_Domestic_Gross'].astype(str).str.replace('\\$', "", regex=True)
```

```python
data.sample()
```

```python
data['USD_Production_Budget'] = data['USD_Production_Budget'].astype(str).str.replace(',', "", regex=True)
data['USD_Worldwide_Gross'] = data['USD_Worldwide_Gross'].astype(str).str.replace(',', "", regex=True)
data['USD_Domestic_Gross'] = data['USD_Domestic_Gross'].astype(str).str.replace(',', "", regex=True)
```

```python
data.USD_Production_Budget = data.USD_Production_Budget.astype('int64')
data.USD_Worldwide_Gross = data.USD_Worldwide_Gross.astype('int64')
data.USD_Domestic_Gross = data.USD_Domestic_Gross.astype('int64')
```

```python
data.sample()
```

```python
data['Release_Date'] = pd.to_datetime(data['Release_Date'], format="%m/%d/%Y")             
```

```python
data.dtypes
```

- The datatypes for columns displaying currency have been fixed, the datatype for the Release_Date column has been changed to datetime

## Data Visualisation 

```python
plt.figure(figsize=(8,4), dpi=200)
with sns.axes_style('darkgrid'):
    ax = sns.scatterplot(data=data, x='USD_Production_Budget', y='USD_Worldwide_Gross', hue='USD_Worldwide_Gross')
    ax.set(
        ylim=(0,1000000000),
        xlim=(0,500000000),
        ylabel='Revenue $ Billions',
        xlabel='Budget $100 Millions'
    )
```

- The majority of movies have a budget of \\$100 Million or less
- Generally movies with a higher budget have resulted in a higher revenue

```python
plt.figure(figsize=(8,4), dpi=200)
with sns.axes_style('darkgrid'):
    ax = sns.scatterplot(
        data=data,
        x='Release_Date',
        y='USD_Production_Budget',
        hue='USD_Worldwide_Gross',
        size='USD_Worldwide_Gross'
    )
    ax.set(
        ylim=(0,450000000),
        xlim=(data.Release_Date.min(), data.Release_Date.max()),
        ylabel='Revenue $ Billions',
        xlabel='Year'
    )
```

- There were significantly more films released from the year 2000 and onwards.
- Movie budgets saw signficiant growth from the 1980s and onwards

```python
old_films = data.loc[data.Decade <= 1969]
old_films.head()
```

```python
new_films = data.loc[data.Decade >= 1970]
new_films.head()
```

```python
old_films.shape
```

```python
new_films.shape
```

```python
old_films.sort_values('USD_Production_Budget', ascending=False).head(1)
```

```python
new_films.sort_values('USD_Production_Budget', ascending=False).head(1)
```

- There are 153 films released before the year 1970
- The most expensive film made prior to 1970 was 'Cleopatra', with a budget of \\$42,000,000
- There are 5229 films released after the year 1970
- The most expensive film made after 1970 was 'Avatar', with a budget of \\$425,000,000

```python
plt.figure(figsize=(8,4), dpi=200)
with sns.axes_style("whitegrid"):
    ax = sns.regplot(
        data=old_films,
        x='USD_Production_Budget',
        y='USD_Worldwide_Gross',
        scatter_kws = {'alpha': 0.5},
        line_kws = {'color': '#ff7c43'}
    )
    ax.set(
    ylabel='Revenue $ Billions',
    xlabel='Budget $ Millions',
    )
```

- Many lower budget films actually made more money than the film with the highest budget

```python
plt.figure(figsize=(8,4), dpi=200)
with sns.axes_style("darkgrid"):
    ax = sns.regplot(
        data=new_films,
        x='USD_Production_Budget',
        y='USD_Worldwide_Gross',
        scatter_kws = {'alpha': 0.3},
        line_kws = {'color': '#ff7c43'}
    )
    ax.set(
        ylabel='Revenue $ Billions',
        xlabel='Budget $ Millions',
    )
```

- Data is much better aligned with the regression line 
- Narrower confidence level indicatesimproved certainty of the relationship between budget and worldwide revenue

```python
regression = LinearRegression()
X = pd.DataFrame(old_films, columns=['USD_Production_Budget'])
y = pd.DataFrame(old_films, columns=['USD_Worldwide_Gross'])
regression.fit(X,y)
```

```python
regression.intercept_
```

```python
regression.coef_
```

```python
regression.score(X,y)
```

- The linear model does not appear to be appropriate for pre-1970s films, the slope of 1.6 indicates a positive relationship between budgets and revenue, however the score of 2% is very low. This makes sense as in the visualisation for old_films, many points were not very well aligned with the regression line.

```python
regression = LinearRegression()
X = pd.DataFrame(new_films, columns=['USD_Production_Budget'])
y = pd.DataFrame(new_films, columns=['USD_Worldwide_Gross'])
regression.fit(X,y)
```

```python
regression.intercept_
```

```python
regression.coef_
```

```python
regression.score(X,y)
```

The slope of 3.12 indicates a positive relationship between budget and revenue, the score of 55% indicates a good fit between the model and data. In the visualisation of new_films, many points were well aligned with the regression line.
